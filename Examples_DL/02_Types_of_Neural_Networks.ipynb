<h1>Types of Neural Networks</h1>
<ul>
  <li>Artificial Neural Networks (ANN)</li>
  <li>Convolutional Neural Networks (CNN)</li>
  <li>Recurrent Neural Networks (ANN)</li>
</ul>

<h5>Artificial Neural Networks (ANN)</h5>

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import plot_model
```

```python
# Define the model architecture
model=Sequential()
```

```python
model.add(Dense(5, input_shape=(3,), activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
# Visualize the model architecture
plot_model(model, show_shapes=True)
```

<h5>Convolutional Neural Networks (CNN)</h5>

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.utils import plot_model
```

```python
# Define the model architecture
model=Sequential()

model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
# Visualize the model architecture
plot_model(model, show_shapes=True)
```

<h5>Recurrent Neural Networks(RNN)</h5>

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from keras.utils import plot_model
```

```python
# Define the model architecture
model=Sequential()

model.add(Embedding(input_dim=1000, output_dim=64))

model.add(LSTM(128))

model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Visualize the model architecture
plot_model(model, show_shapes=True)
```
