<h5>How to load a video file<h5>

```python
source=cv2.VideoCapture('sample/video_file.mp4')
#0-webcam
#1,2,3=usb camera
#myVideo.mp4-file
#124.35.32.43.643 p/w user name - wifi camera

ret, img= source.read()
print(ret)
cv2.imshow('IMG',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
source=cv2.VideoCapture('sample/video_file.mp4')
while(True):
    ret, img=source.read()
    cv2.rectangle(img,(20,20),(100,100),(0,255,0),-1)
    if(ret==False):
        break

    cv2.imshow('IMG',img)
    key=cv2.waitKey(1)

    if(key==27):
        break

cv2.destroyAllWindows()
```

<h5>Image Thresholding</h5>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 

img=cv2.imread('sample/threshold.jpg')
plt.imshow(img,cmap='gray')
```

```python
ret, thresh=cv2.threshold(img,60,255,cv2.THRESH_BINARY)
plt.imshow(thresh,cmap='gray')
```

```python
ret, thresh=cv2.threshold(img,60,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh,cmap='gray')
```

```python
ret, thresh=cv2.threshold(img,60,255,cv2.THRESH_TOZERO)
plt.imshow(thresh,cmap='gray')
```

```python
ret, thresh=cv2.threshold(img,60,255,cv2.THRESH_TRUNC)
plt.imshow(thresh,cmap='gray')
```
